### Data Preparation & Cleaning
##### Modifying column names for ease of access. 

```python
df.columns = df.columns.map(lambda x: x.replace(" ","_"))
```

```python
df.select_dtypes("object").columns
```

```python
df.Holiday_Dummy.value_counts()
```

##### '0' represents 'No Holiday', '1' represents 'Holiday'

```python
df.Holiday_Dummy = df.Holiday_Dummy.apply(lambda x : 0 if (x := x.strip()) == "-" else 1)
df.Holiday_Dummy.value_counts()
```

#####  Converting Week_Ending to datetime column

```python
df.Week_Ending = pd.to_datetime(df.Week_Ending, format="%d-%m-%Y %H:%M")
```

##### Handling columns like Paid Search Impressions, Paid Search Spends

```python
import numpy as np
df.Paid_Search_Impressions = df.Paid_Search_Impressions.apply(lambda x: np.NaN if x.strip() == "-" else x)
df.Paid_Search_Spends =  df.Paid_Search_Spends.apply(lambda x: np.NaN if x.strip() == "-" else x)
```

##### Handling Datatypes

```python
cols = ['Sales','Total_SKU', 'Gasoline_Price', 'Email_Clicks', 'Email_Spends',
       'Organic_Search_Impressions', 'Modular_Video_Impressions',
       'Modular_Video_Spends', 'Paid_Social_Impressions', 'Paid_Social_Spends',
       'Paid_Search_Impressions','Paid_Search_Spends']

for i in cols:
    df[i] = df[i].str.replace(",","") 
    df[i] = df[i].astype("float")
```

### Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as w

w.filterwarnings("ignore")
sns.set_theme()
```

### Plot sales trend over time

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x=df["Week_Ending"], y=df["Sales"], marker="o", color="b")
plt.title("Weekly Sales Trend")
plt.xlabel("Week Ending")
plt.ylabel("Sales")
plt.xticks(rotation=40)
plt.show()
```

#### The sales trend shows fluctuations over time, with possible seasonal peaks.

### Find the correlations between sales and different marketing and external factors. 

```python
c_matrix = df.corr()
c_m = c_matrix[["Sales"]].sort_values(by="Sales", ascending=False)
```

#### Ploting heatmap of correlations with Sales.

```python
plt.figure(figsize=(6,7))
sns.heatmap(c_m, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation of Features with Sales")
plt.show()
```

### Seasonality patterns to determine if sales fluctuate periodically.
#### Plotting monthly sales trend.

```python
df["Month"] = df["Week_Ending"].dt.month
monthly_sales = df.groupby("Month")["Sales"].mean()

plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette="Blues_r")
plt.title("Average Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()
```
