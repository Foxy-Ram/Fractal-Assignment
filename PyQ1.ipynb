```python
# Relevant Imports
import pandas as pd

# Data Preparation 
data1 = [["21-07-2021 00:00", 129420], 
         ["28-07-2021 00:00", 150990], 
         ["04-08-2021 00:00",107850],
         ["11-08-2021 00:00", 53925],
         ["18-08-2021 00:00", 215700], 
         ["25-08-2021 00:00", 291195],
         ["01-09-2021 00:00", 107850],
         ["08-09-2021 00:00", 312765], 
         ["15-09-2021 00:00", 431400],
         ["22-09-2021 00:00", 97065],
         ["29-09-2021 00:00", 204915],
         ["06-10-2021 00:00", 215700],
         ["13-10-2021 00:00", 53925]]
            
data2 = [["04-08-2021 00:00", 5],
         ["11-08-2021 00:00", 17],
         ["18-08-2021 00:00", 25],
         ["25-08-2021 00:00", 29],
         ["01-09-2021 00:00", 23]]

data3 = [["21-07-2021 00:00", 2500],
         ["28-07-2021 00:00", 5000],
         ["04-08-2021 00:00", 10000],
         ["11-08-2021 00:00", 7500],
         ["18-08-2021 00:00", 11250],
         ["25-08-2021 00:00", 7000],
         ["01-09-2021 00:00", 5750]]



cols1 = ["Timestamp", "Sales"]
cols2 = ["Timestamp", "TV GRPs"]
cols3 = ["Timestamp", "Video Impressions"] 

df1 = pd.DataFrame(data=data1, columns=cols1)
df2 = pd.DataFrame(data=data2, columns=cols2)
df3 = pd.DataFrame(data=data3, columns=cols3)
```

```python
data4 = {"Timestamp" : ["21-07-2021 00:00", "28-07-2021 00:00", "04-08-2021 00:00", "11-08-2021 00:00", "18-08-2021 00:00",
              "25-08-2021 00:00", "01-09-2021 00:00", "08-09-2021 00:00", "15-09-2021 00:00", "22-09-2021 00:00",
              "29-09-2021 00:00", "06-10-2021 00:00", "13-10-2021 00:00"]}

df4 = pd.DataFrame(data4)
```

### Q. Merge the three data sets  with sales, TV GRPs & Video Impressions

### Solution

```python
# Merge operation
df4 = df4.merge(df1,on="Timestamp",how="left")\
         .merge(df2, on="Timestamp", how="left")\
         .merge(df3, on="Timestamp",how="left")
df4
```
